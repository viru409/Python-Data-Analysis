```python
import pandas as pd
import numpy as np
```

```python
 data = pd.Series(np.random.randn(9),
                  index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                         [1, 2, 3, 1, 3, 1, 2, 2, 3]])
print(data)
print(data.index)
```

```python
# Hierarchical indexing object or Partial indexing
print(data['b'])
print()
print(data['b' : 'c'])
print()
print(data.loc[['b' , 'c']])
print()
print(data.loc[:, 2])
```

```python
# Arrange the series data in dataframe
data.unstack()
```

```python
# Inverse of unstack operation is stack
data.unstack().stack()
```

```python
frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                     columns=[['Ohio', 'Ohio', 'Colorado'],
                                              ['Green', 'Red', 'Green']])
print(frame)
frame.index.names = ['key1' , 'key2']
frame.columns.names = ['state', 'color']
print()
print(frame)
```

```python
frame['Ohio'], frame['Colorado']
```

```python
print(frame)
print(frame.index)
```

```python
# Reordering & Sorting levels

frame.swaplevel('key1', 'key2')
```

```python
frame.swaplevel(0, 1).sort_index(level=0)
```

```python
# aggregate level on either the rows or columns
frame.sum(level='key2')   
```

```python
frame.sum(level='color', axis=1)
```

```python
# Indexing with DataFrame columns

frame = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1),
                      'c': ['one', 'one', 'one', 'two', 'two',
                            'two', 'two'],
                      'd': [0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
frame2 = frame.set_index(['c','d'])
frame2
```

```python
frame.set_index(['c', 'd'], drop=False) # By default the columns are removed from the DataFrame, though you can leave them
```
